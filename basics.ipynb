### Problem Statement

Given an array `nums` of `n` integers, find the most frequent element in it â€” i.e., the element that occurs the maximum number of times.
If there are multiple elements that appear the maximum number of times, return the **smallest** among them.

### Examples

#### Example 1
**Input:**
`nums = [1, 2, 2, 3, 3, 3]`

**Output:**
`3`

**Explanation:**
The number `3` appears **3 times**, which is more than any other number.
Hence, `3` is the most frequent element.

---

#### Example 2
**Input:**
`nums = [4, 4, 5, 5, 6]`

**Output:**
`4`

**Explanation:**
Both `4` and `5` appear twice, but since `4` is **smaller**, we choose `4`.

---


```python
nums = [1, 2, 2, 3, 3, 3, 4, 4, 4]

def mostFrequentElement(nums):
    freq = {}
    for num in nums:
        if num in freq:
            freq[num] += 1
        else:
            freq[num] = 1
    # print(freq)
    # print(f"Key is: {min(k for k, v in list(freq.items()) if v == max(v for k, v in list(freq.items())))}")
    return min(k for k, v in list(freq.items()) if v == max(v for k, v in list(freq.items())))

mostFrequentElement(nums)
```
