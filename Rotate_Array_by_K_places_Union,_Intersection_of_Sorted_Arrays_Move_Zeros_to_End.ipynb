Given an array `arr` and an integer `k`, rotate the array **to the right by k positions**.
Rotation means each element is shifted to the right, and the last element wraps around to the beginning.

Example:
`arr = [1, 2, 3, 4, 5]`, `k = 2` â†’ `[4, 5, 1, 2, 3]`

## ğŸ§¨ 1. Brute Force Approach (Using Extra Space)

Rotate the array by using a temporary array.

### âš™ï¸ Logic

1. Create a temporary array of the same size.
2. Place each element at its correct rotated index:
   `new_index = (i + k) % n`
3. Copy the temp array back to the original.

### ğŸ§  Pseudocode

```
procedure rotateBrute(arr, k)
    n = length(arr)
    temp = new array of size n

    for i = 0 to n - 1 do
        temp[(i + k) % n] = arr[i]
    end for

    for i = 0 to n - 1 do
        arr[i] = temp[i]
    end for
end procedure
```

### â±ï¸ Time Complexity
- **O(n)**

### ğŸ’¾ Space Complexity
- **O(n)** (extra array)

---

## ğŸš€ 2. Better Approach (Rotate One by One)

Rotate the array right by 1, **k times**.

### âš™ï¸ Logic

1. For each rotation:
   - Store the last element.
   - Shift all elements to the right by 1.
   - Place saved element at index 0.

### ğŸ§  Pseudocode

```
procedure rotateBetter(arr, k)
    n = length(arr)
    for r = 1 to k do
        last = arr[n - 1]
        for i = n - 1 down to 1 do
            arr[i] = arr[i - 1]
        end for
        arr[0] = last
    end for
end procedure
```

### â±ï¸ Time Complexity
- **O(n Ã— k)** â†’ inefficient when k is large

### ğŸ’¾ Space Complexity
- **O(1)**

---

## âš¡ 3. Optimal Approach â€” Reversal Algorithm

This is the **best and most used** method.

### âš™ï¸ Logic

To rotate right by `k`:

1. Reverse the entire array
2. Reverse the first `k` elements
3. Reverse the remaining `n - k` elements

This uses the fact that reversing segments places them in correct rotated positions.

### ğŸ§  Pseudocode

```
procedure reverse(arr, start, end)
    while start < end do
        swap arr[start] and arr[end]
        start = start + 1
        end = end - 1
    end while
end procedure


procedure rotateOptimal(arr, k)
    n = length(arr)
    k = k % n

    reverse(arr, 0, n - 1)
    reverse(arr, 0, k - 1)
    reverse(arr, k, n - 1)
end procedure
```

### â±ï¸ Time Complexity
- **O(n)**

### ğŸ’¾ Space Complexity
- **O(1)** (in-place)

---

**Input:**
`arr = [1, 2, 3, 4, 5]`, `k = 2`

**Step 1:** Reverse all â†’ `[5, 4, 3, 2, 1]`
**Step 2:** Reverse first 2 â†’ `[4, 5, 3, 2, 1]`
**Step 3:** Reverse rest â†’ `[4, 5, 1, 2, 3]`

**Output:**
`[4, 5, 1, 2, 3]`

# ğŸ”„ Move All Zeroes to the End of the Array

### ğŸ“˜ Introduction

The goal is to **move all zeroes** in the array to the **end**, while maintaining the **relative order of the non-zero elements**.
This must be done **in-place** without using extra space for another array.

---

### âš™ï¸ How It Works

1. Maintain a pointer `i` that tracks the position to place the next non-zero element.
2. Traverse the array:
   - When you find a non-zero element, place it at index `i` and increment `i`.
3. After processing all elements, fill the remaining positions from `i` to the end with **zeroes**.
4. The array ends up with all non-zero elements in front (in order), followed by all zeroes.

---

### ğŸ§  Pseudocode

```
procedure moveZeroes(arr)
    n = length(arr)
    i = 0

    for j = 0 to n - 1 do
        if arr[j] â‰  0 then
            arr[i] = arr[j]
            i = i + 1
        end if
    end for

    while i < n do
        arr[i] = 0
        i = i + 1
    end while
end procedure
```

---

**Input:**
`arr = [0, 1, 0, 3, 12]`

**Process:**
- Move non-zeroes: `[1, 3, 12, _, _]`
- Fill remaining with zeroes â†’ `[1, 3, 12, 0, 0]`

**Output:**

### ğŸ“˜ Introduction

Given two **sorted arrays**, the goal is to find their **union** â€”
i.e., all **unique elements** that appear in either of the arrays.

Since both arrays are already sorted, we can use a **two-pointer approach** to efficiently gather unique elements in ascending order.

---

### âš™ï¸ How It Works

1. Initialize two pointers `i` and `j` at the start of both arrays.
2. Compare elements:
   - If `arr1[i] < arr2[j]`, add `arr1[i]` to the result and move `i`.
   - If `arr2[j] < arr1[i]`, add `arr2[j]` to the result and move `j`.
   - If both are equal, add the element once and move **both** pointers.
3. Skip duplicates by checking the **last inserted element**.
4. When one array is exhausted, add remaining elements from the other (ignoring duplicates).
5. The result contains the union in sorted order.

---

### ğŸ§  Pseudocode

```

                j = j + 1
            else

                i = i + 1

            end if

```
