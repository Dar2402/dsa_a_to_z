### ğŸ“˜ Introduction

Given an array of integers, the task is to find the **second largest element** in the array.
You must do this **without sorting** the entire array.

If there are duplicates of the largest element, the second largest should still be **distinct** from it.

---

### âš™ï¸ How It Works

1. Initialize two variables â€” `largest` and `second_largest`.
2. Traverse the array:
   - If the current element is **greater than `largest`**, update `second_largest = largest` and `largest = current_element`.
   - Else if the current element is **less than `largest` but greater than `second_largest`**, update `second_largest = current_element`.
3. Continue until the end of the array.
4. Return `second_largest`.

---

### ğŸ§  Pseudocode

```

    if length(arr) < 2 then
        return "Not enough elements"

    largest = -âˆ
    second_largest = -âˆ

    for i = 0 to length(arr) - 1 do
        if arr[i] > largest then
            second_largest = largest

        else if arr[i] > second_largest and arr[i] != largest then

        end if
    end for

```

````python
class Solution:
    def secondLargestElement(self, nums):
        if len(nums) < 2:
            return -1
        largest = float('-inf')
        s_largest = float('-inf')

        for i, v in enumerate(nums):
            if nums[i] > largest:
                s_largest = largest
                largest = nums[i]
            elif nums[i] > s_largest and nums[i] < largest:
                s_largest = nums[i]
        return s_largest if s_largest != float('-inf') else -1

s = Solution()
s.secondLargestElement([10, 10, 10])
````

### ğŸ“˜ Introduction

Given an array of integers, the task is to determine whether the array is **sorted in non-decreasing order** (i.e., every element is **greater than or equal to** the previous one).

This is a common and fundamental check used in searching, sorting, and validation logic.

---

### âš™ï¸ How It Works

To verify if the array is sorted:

1. Traverse the array from left to right.
2. Compare each element with the previous one.
3. If any element is **smaller** than its predecessor, the array is **not sorted**.
4. If no such violation is found, the array is **sorted**.

---

### ğŸ§  Pseudocode

```
procedure isSorted(arr)
    n = length(arr)

    for i = 1 to n - 1 do
        if arr[i] < arr[i - 1] then
            return false
        end if
    end for

    return true
end procedure
```

---

### ğŸ’» Example

**Input:**
`arr = [1, 2, 2, 3, 5, 8]`

**Output:**
`true`

**Explanation:**
Every element is greater than or equal to the previous element.

---

### âŒ Example (Not Sorted)

**Input:**
`arr = [3, 5, 4, 8, 10]`

**Output:**
`false`

**Explanation:**
The element `4` violates the sorted order because `4 < 5`.

| `[7]` | true | Single element is trivially sorted |

### ğŸ“˜ Introduction

Given a **sorted array**, the task is to **remove duplicates in-place** so that each element appears **only once**.
The function should return the **number of unique elements**, and the array should be modified so that the **first k elements** contain those unique values.

(After `k` elements, the remaining values donâ€™t matter.)

This problem is commonly asked in interviews because it tests understanding of **two-pointer techniques**.

---

### âš™ï¸ How It Works

1. Use two pointers:
   - `i` â†’ tracks the *position of the last unique element*.
   - `j` â†’ scans through the array.
2. When `arr[j]` is different from `arr[i]`,
   - increment `i`,
   - update `arr[i] = arr[j]`.
3. Continue until the end.
4. The final unique count is `i + 1`.

---

### ğŸ§  Pseudocode

```
procedure removeDuplicates(arr)
    n = length(arr)
    if n == 0 then
        return 0

    i = 0
    for j = 1 to n - 1 do
        if arr[j] != arr[i] then
            i = i + 1
            arr[i] = arr[j]
        end if
    end for

    return i + 1   // number of unique elements
end procedure
```

- Encounter `4` â†’ `[1, 2, 3, 4]`

**Output:**
- Unique Count â†’ `4`
